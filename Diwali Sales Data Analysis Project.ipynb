Exploratory Data Analysis (EDA)

```python
#Count the number of Male & Women by Purchasing
sns.countplot(data=df , x="Gender", hue="Gender" )
plt.title("Men & Women Comparision by Purchase",fontsize=20)
plt.grid()
plt.show()
```

```python
ax=sns.countplot(x="Gender" , data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
df['Gender'].count()
```

```python
# The Total Amount which has been paid by Male & Female
data= df.groupby(['Gender'],as_index=False)['Amount'].sum()
data
```

```python
sns.barplot(data=data, x="Gender",y="Amount", hue="Gender" )
plt.show()
```

From above graphs we can see that females are consider that the most buyers and even the purchasing power of females are greater than men

```python
#Which a specific Age Group is concider the most buyers
ax= sns.countplot(data= df, x = "Age Group", hue="Gender")
for bars in ax.containers :
    ax.bar_label(bars)
    plt.show()
```

```python
#The Total Amount Which Has Been Spent Of Each Age Group
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sales_age
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
total=sum(sales_age['Amount'])
for i, v in enumerate(sales_age["Amount"]):
    percentage = v/total *100
    plt.text(i, v, f'{percentage:.2f}%', ha="center", va="bottom")
    plt.show()
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female 

```python
# The Total Orders according to each State 
orders_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
orders_state
```

```python
plt.figure(figsize=(17,8))
sns.barplot(data=orders_state , x='State',y='Orders',hue='State')
plt.show()
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

Married

```python
ax= sns.countplot(data=df, x='Married',hue='Married')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
sales_by_gender_and_married= df.groupby(['Gender','Married'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_by_gender_and_married
```

```python
sns.set(rc={'figure.figsize':(15,6)})
ax=sns.barplot(data=sales_by_gender_and_married, x='Married', y='Amount' ,hue='Gender')
total = sum(sales_by_gender_and_married['Amount'])
for i, v in enumerate(sales_by_gender_and_married["Amount"]):
    percentage = v/total *100
    plt.text(i, v, f'{percentage:.2f}%', ha="center", va="bottom")
    plt.show()
```

From above Graphs we could see that most of buyers are not married (females) and they have high purchasing power

```python
# Analysis Sales by Occupation
sales_occupation =  df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by="Amount" , ascending=False)
sales_occupation 
```

```python
sns.set(rc={'figure.figsize':(20,7)})
ax= sns.countplot(data=df,x='Occupation',hue='Occupation',legend=True)
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,7)})
ax=sns.barplot(data=sales_occupation , x='Occupation' , y='Amount', hue='Occupation',legend=True)
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

```python
# Analysis Sales by Product Category
sales_by_product_category= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False ).head(10)
sales_by_product_category
```

```python
sns.set(rc={'figure.figsize':(25,8)})
sns.countplot(data=df, x= 'Product_Category', hue='Product_Category')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(25,15)})
ax=sns.barplot(data=sales_by_product_category, x='Product_Category', y='Amount',hue='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics categories
